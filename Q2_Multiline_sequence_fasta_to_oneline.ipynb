```python
def reformed_fasta(input_file, output_file):
  try:
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        sequence=[]
        for line in infile:
          line=line.strip()
          if line.startswith(">"):
            if sequence:
              outfile.write("".join(sequence) + "\n")
              sequence.clear()
            outfile.write(line + "\n")
          else:
              sequence.append(line)

        if sequence:
            outfile.write("".join(sequence) + "\n")

    print("Conversion complete. Output saved to :", outfile)

  except FileNotFoundError:
    print(" File not found: ", infile)
  except Exception as e:
    print("Error occured: ", e)


def main():
  input_fasta= 'multiline_input.fasta'
  output_fasta= 'output.fasta'

  reformed_fasta(input_fasta, output_fasta)

if __name__=="__main__":
  main()
```
