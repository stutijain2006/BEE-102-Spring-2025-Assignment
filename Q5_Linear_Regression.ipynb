```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Let the true model be y= 3x+7
np.random.seed(42)
x= np.array( [1,2,3,4,5,6]).reshape(-1,1)
true_y= 3*x +7
noise= np.random.normal( 0,2, size=x.shape)
y=( true_y + noise).flatten()

model= LinearRegression()
model.fit( x,y )
y_pred= model.predict(x)

y_mean= np.mean(y)
total_sum_of_squares= np.sum((y-y_mean**2))
explained_sum_of_squares = np.sum((y_pred-y_mean)**2)
residual_sum_of_squares= np.sum((y-y_pred)**2)

print("Total Sum of squares: ", total_sum_of_squares)
print("Explained Sum of Squares: ", explained_sum_of_squares)
print("Residual Sum of Squares: ", residual_sum_of_squares)
print("Total sum of squares= Explained sum of squares + Residual sum of squares?", np.isclose(total_sum_of_squares, (explained_sum_of_squares + residual_sum_of_squares)))

plt.scatter(x,y, label='Observed', color='blue')
plt.plot(x, y_pred, label='Predicted(Regression Line)', color='red')
plt.hlines(y_mean, xmin= x.min(), xmax= x.max(), colors='green', label='Mean of y')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
